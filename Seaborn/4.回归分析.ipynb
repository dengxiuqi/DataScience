```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

#### 4.1 绘制回归线
regplot()和lmplot()都可以绘制回归关系,推荐regplot()

```python
sns.regplot(x="total_bill", y="tip", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips);
```

#### 4.2 离散数据

```python
sns.regplot(data=tips,x="size",y="tip")
```

对某些离散化数据，出现这种情况时（数据点有重叠），可以使用一个参数设定解决。

```python
sns.regplot(x="size", y="tip", data=tips, x_jitter=0.1)
```

#### 4.3 特征内部分组对比

lmplot的hue参数按cloumns分组，一图多组

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

lmplot的hue参数按cloumns分组，加col参数——列增加，加row参数——行增加

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", row="sex", data=tips, markers=["o", "x"], size=3)
```

lmplot的col参数分出来多组后，col_wrap参数控制图形显示排布；size参数设置图形大小，aspect参数设置图形横纵比

```python
sns.lmplot(x="total_bill", y="tip", col="day", data=tips, col_wrap=2, size=3, aspect=0.8)#
```

#### 4.4 设置数据点的形状和颜色

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, markers=["o", "x"], palette="Set1");
```
