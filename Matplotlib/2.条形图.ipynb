```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = ["a","b","c","d","e","f"]
y = [56.01,26.94,17.53,16.49,15.45,12.96]
```

#### 2.1 bar绘制
        left:长条形中点横坐标
        height:长条形高度
        width:长条形宽度，默认值0.8
        label:为后面设置legend准备

```python
plt.figure(figsize = (8, 5))
plt.bar(x=range(len(y)), height=y, width=0.5)
```

#### 2.2 刻度
* 在matplotlib中ticks表示的是刻度，而刻度有两层意思，一个是刻标(locs)，一个是刻度标签(tick labels)
* 在作图时，x轴y轴都是连续的，所以刻标可以随意指定,就是在连续变量上找寻位置，而刻度标签则可以对应替换



```python
plt.figure(figsize = (8, 5))
plt.bar(x=range(len(y)), height=y, width=0.5, color="orange")

plt.xticks(range(len(y)), x, fontproperties='SimHei')
plt.show()
```

#### 2.3 条形图高度的标注

```python
plt.figure(figsize = (8, 5))
rects = plt.bar(x=range(len(y)), height=y, width=0.5, color="g")
plt.xticks(range(len(y)), x)

for rect in rects:
        height = rect.get_height()
        plt.text(x=rect.get_x() + rect.get_width()/2, 
                 y=height+1, 
                 s=height, 
                 ha="center", va="bottom")
plt.show()
```

#### 2.4 显示图例

```python
plt.figure(figsize = (8, 5))
plt.bar(x=x, height=y, width=0.5, label="Jacobi")

plt.legend()
```

#### 2.5 横条图

```python
plt.barh(x, y, color="m")
```

#### 2.6 连续值填充

```python
fig,ax = plt.subplots()
ax.fill_between(x=np.random.randn(100).cumsum(),
                y1=np.linspace(0,10,100),
                color='lightblue')
plt.show()
```

#### 2.7 误差棒

```python
error = [np.math.sqrt(h) for h in y]

plt.bar(x, y, yerr=error, alpha=0.5)
plt.show()
```

#### 2.8 双向对比

```python
y2 = [-h+np.random.randint(-20, 20) for h in y]

plt.barh(x, y, color='g',alpha = 0.5)
plt.barh(x, y2,color='b',alpha = 0.5)
```

#### 2.9 多种数据的条形图

```python
green_data = [1, 2, 3]
blue_data = [3, 2, 1]
red_data = [2, 3, 3]
labels = ['group 1', 'group 2', 'group 3']
```

```python
pos = list(range(len(green_data))) 
width = 0.2 
plt.bar(pos,green_data,width,alpha=0.5,color='g',label=labels[0])
plt.bar([p+width for p in pos], blue_data,width,alpha=0.5,color='b',label=labels[1])
plt.bar([p+width*2 for p in pos], red_data,width,alpha=0.5,color='r',label=labels[2])
plt.legend()
```

#### 2.10 基准线

```python
mean = np.mean(y)
bars = plt.barh(x, y, alpha = 0.5, color='g')
plt.vlines(mean, -1, len(y),  linestyle = 'dashed')
for b,d in zip(bars, y):
    plt.text(b.get_width()+b.get_width()*0.05,b.get_y()+b.get_height()/2,'{0:.2%}'.format(d/mean))
```
