```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Scatter参数说明
            x，y：array_like，shape（n，） 输入数据
            s：标量或array_like，shape（n，），可选, 大小以点数^ 2。默认是`rcParams ['lines.markersize'] ** 2`。
            c：颜色，顺序或颜色顺序，可选，默认：'b'
            marker：可选，默认值：'o'
            cmap：可选，默认：无
            norm：`可选，默认 `normalize`。
            vmin，vmax：标量，可选，默认值：无; `vmin`和`vmax`与`norm`结合使用来标准化; 亮度数据。
            alpha：混合值，介于0（透明）和1（不透明）之间，
            linewidths：标量或array_like，可选，默认值：无, 如果无，则默认为（lines.linewidth，）。


#### 4.1 散点图

```python
x = np.random.uniform(-10, 10, 100)
y = np.random.uniform(-10, 10, 100)
plt.scatter(x, y)
```

#### 4.2 改变点的形状

```python
x1 = np.random.normal(-50, 10, 100)
y1 = np.random.normal(0, 10, 100)

x2 = np.random.normal(30, 20, 100)
y2 = np.random.normal(0, 20, 100)

x3 = np.random.normal(0, 50, 100)
y3 = np.random.normal(0, 50, 100)
```

```python
plt.scatter(x1, y1,marker ='x', color='blue',alpha=0.6,label='x1')
plt.scatter(x2, y2, marker ='o', color='red',alpha=0.6,label='x2')
plt.scatter(x3, y3, marker ='^', color='green',alpha=0.6,label='x3')
plt.legend(loc='best')
```

#### 4.4 改变点的大小

```python
mu_vec = np.array([0,0])
cov_mat = np.array([[1,0],[0,1]])
X = np.random.multivariate_normal(mu_vec, cov_mat, 500)
```

```python
R=X**2
R_sum=R.sum(axis=1)
plt.scatter(X[:,0], X[:,1], color='green', marker='o', s=32.* R_sum, edgecolor='black', alpha=0.5)
```

#### 4.5 标记点的坐标

```python
x = [0.13, 0.22, 0.39, 0.59, 0.68, 0.74, 0.93]
y = [0.75, 0.34, 0.44, 0.52, 0.80, 0.25, 0.55]

plt.scatter(x, y, marker='s', s=50)

for i, j in zip(x, y):
    plt.annotate('(%s,%s)'%(i, j), xy=(i, j), xytext=(0,-15), textcoords = 'offset points', ha='center')
```
