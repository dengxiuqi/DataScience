```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

#### 可选风格

```python
plt.style.available
```

#### 正常风格

```python
x = np.linspace(-10,10)
y = np.sin(x)
plt.plot(x,y)
```

#### 各种风格

```python
plt.style.use('dark_background')
plt.plot(x,y)
```

```python
plt.style.use('bmh')
plt.plot(x,y)
```

```python
plt.style.use('ggplot')
plt.plot(x,y)
```

```python
plt.style.use(['ggplot','bmh'])
plt.plot(x,y)
```

```python
plt.xkcd()
plt.plot(x,y)
```
