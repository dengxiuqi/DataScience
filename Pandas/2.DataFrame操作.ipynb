#### 2.9 映射函数

```python
data = pd.DataFrame({'food':['A1','A2','B1','B2','B3','C1','C2'], 'data':[1,2,3,4,5,6,7]})
data
```

```python
def foo(series):
    if series['food']=='A1' or series['food']=='A2':
        return 'A'
    elif series['food']=='B1' or series['food']=='B2' or series['food']=='B3':
        return 'B'
    elif series['food']=='C1' or series['food']=='C2':
        return 'C'
```

```python
data['food_map']  = data.apply(foo, axis=1)
data
```

#### 2.10 映射字典

```python
food2Upper = {
    'A1':'A',
    'A2':'A',
    'B1':'B',
    'B2':'B',
    'B3':'B',
    'C1':'C',
    'C2':'C',
}
```

```python
data['upper']=data['food'].map(food2Upper)
data
```
